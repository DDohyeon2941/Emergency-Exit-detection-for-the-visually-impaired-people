Augment images for each class

```python
trn_dir = '/content/gdrive/My Drive/IOT_Project/data/Triple/Train/Raw'
trn_aug_dir = '/content/gdrive/My Drive/IOT_Project/data/Triple/Train/Augmented'
trn_cls_0_path = os.path.join(trn_dir, 'Exit_arrow'), os.path.join(trn_aug_dir, 'Exit_arrow')
trn_cls_1_path = os.path.join(trn_dir, 'Exit_here'), os.path.join(trn_aug_dir, 'Exit_here')
```

```python
temp_gen = ImageDataGenerator(horizontal_flip=True, vertical_flip=True, brightness_range=[0.7,1.3])
temp_flow_gen = temp_gen.flow_from_directory(trn_cls_0_path[0], shuffle = False, target_size=(224, 224), batch_size=1, class_mode = None, save_to_dir = trn_cls_0_path[1])
temp_flow_gen1 = temp_gen.flow_from_directory(trn_cls_1_path[0], shuffle = False, target_size=(224, 224), batch_size=1, class_mode = None, save_to_dir = trn_cls_1_path[1])
```

Class 0, Exit with Arrow

```python
added_size=0
for uni_img in temp_flow_gen:
    if added_size > 1000:
        break
    if added_size % 25 == 0:
        print(added_size)
    added_size+=1
```

Class 1, Exit without arrow

```python
added_size=0
for uni_img in temp_flow_gen1:
    if added_size > 1000:
        break
    if added_size % 25 == 0:
        print(added_size)
    added_size+=1
```
